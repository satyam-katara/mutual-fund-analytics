```python
import pandas as pd
import os

DATA_PATH = r"D:\projects\mutual-fund-analytics\data\raw"

def load_csv_files():
    files = [f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]
    
    datasets = {}
    
    for file in files:
        path = os.path.join(DATA_PATH, file)
        df = pd.read_csv(path)
        
        print(f"\n📁 File: {file}")
        print("Shape:", df.shape)
        print("Dtypes:\n", df.dtypes)
        print("Head:\n", df.head())
        
        datasets[file] = df
    
    return datasets

if __name__ == "__main__":
    load_csv_files()
```

```python
import requests
import pandas as pd
import os

SAVE_PATH =  r"D:\projects\mutual-fund-analytics\data\raw"


def fetch_nav(scheme_code):
    url = f"https://api.mfapi.in/mf/{scheme_code}"
    
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"Failed for {scheme_code}")
        return
    
    data = response.json()
    
    nav_data = data['data']
    
    df = pd.DataFrame(nav_data)
    
    file_name = f"{scheme_code}_nav.csv"
    df.to_csv(os.path.join(SAVE_PATH, file_name), index=False)
    
    print(f"Saved {file_name}")

if __name__ == "__main__":
    
    schemes = {
        "HDFC Top 100": 125497,
        "SBI Bluechip": 119551,
        "ICICI Bluechip": 120503,
        "Nippon Large Cap": 118632,
        "Axis Bluechip": 119092,
        "Kotak Bluechip": 120841
    }
    
    for name, code in schemes.items():
        print(f"Fetching {name}")
        fetch_nav(code)
```

```python
import requests
import pandas as pd
import os

scheme_codes = {
    "hdfc_top100": 125497,
    "sbi_bluechip": 119551,
    "icici_bluechip": 120503,
    "nippon_largecap": 118632,
    "axis_bluechip": 119092,
    "kotak_bluechip": 120841
}

save_path = r"D:\projects\mutual-fund-analytics\data\raw"

for name, code in scheme_codes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    data = requests.get(url).json()

    df = pd.DataFrame(data["data"])

    file_name = os.path.join(
        save_path,
        f"{name}_live_nav.csv"
    )

    df.to_csv(file_name, index=False)

    print(f"Saved: {name}")
```
