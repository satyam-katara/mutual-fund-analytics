```python
import os
import pandas as pd

RAW_PATH = r"D:\projects\mutual-fund-analytics\data\raw"
PROCESSED_PATH = r"D:\projects\mutual-fund-analytics\data\processed"
```

```python
import pandas as pd
import os

PROCESSED_PATH = r"D:\projects\mutual-fund-analytics\data\processed"

fund = pd.read_csv(os.path.join(PROCESSED_PATH, "01_fund_master_clean.csv"))
nav = pd.read_csv(os.path.join(PROCESSED_PATH, "02_nav_history_clean.csv"))
transactions = pd.read_csv(os.path.join(PROCESSED_PATH, "08_investor_transactions_clean.csv"))
performance = pd.read_csv(os.path.join(PROCESSED_PATH, "07_scheme_performance_clean.csv"))
aum = pd.read_csv(os.path.join(RAW_PATH, "03_aum_by_fund_house.csv"))
```

```python
fund.to_sql(
    "dim_fund",
    engine,
    if_exists="replace",
    index=False
)

nav.to_sql(
    "fact_nav",
    engine,
    if_exists="replace",
    index=False
)

transactions.to_sql(
    "fact_transactions",
    engine,
    if_exists="replace",
    index=False
)

performance.to_sql(
    "fact_performance",
    engine,
    if_exists="replace",
    index=False
)

aum.to_sql(
    "fact_aum",
    engine,
    if_exists="replace",
    index=False
)

print("All tables loaded successfully!")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    tables = conn.execute(
        text("SELECT name FROM sqlite_master WHERE type='table';")
    )

    for table in tables:
        print(table[0])
```

```python
with engine.connect() as conn:

    table_names = [
        "dim_fund",
        "fact_nav",
        "fact_transactions",
        "fact_performance",
        "fact_aum"
    ]

    for table in table_names:
        result = conn.execute(
            text(f"SELECT COUNT(*) FROM {table}")
        )

        count = result.scalar()

        print(f"{table}: {count} rows")
```
