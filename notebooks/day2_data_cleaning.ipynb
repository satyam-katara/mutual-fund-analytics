```python
performance = pd.read_csv(
    os.path.join(RAW_PATH, "07_scheme_performance.csv")
)

performance.head()
```

```python
performance.info()
```

```python
performance.columns
```

```python
performance.isnull().sum()
```

```python
performance.info()
```

```python
return_columns = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "benchmark_3yr_pct"
]

for col in return_columns:
    performance[col] = pd.to_numeric(performance[col], errors="coerce")

print(performance[return_columns].dtypes)
```

```python
performance[return_columns].isnull().sum()
```

```python
invalid_expense = performance[
    (performance["expense_ratio_pct"] < 0.1) |
    (performance["expense_ratio_pct"] > 2.5)
]

print("Invalid Expense Ratios:", len(invalid_expense))
```

```python
anomalies = performance[
    (performance["return_1yr_pct"] < -100) |
    (performance["return_1yr_pct"] > 100)
]

print("Anomalies Found:", len(anomalies))
```

```python
duplicates = performance.duplicated().sum()

print("Duplicate Rows:", duplicates)

performance = performance.drop_duplicates()
```

```python
performance.to_csv(
    os.path.join(PROCESSED_PATH, "07_scheme_performance_clean.csv"),
    index=False
)

print("Scheme Performance cleaned successfully!")
```
